### The Incremental Load
##### On the terminal :
Open a .sh file and write a script that make the incremental load
    ```
    _4_sqoop_incremental_load.sh
    ```
There is the following script: 
```
#!/bin/bash

### This works and does the incremental load 

sqoop import \
    --connect jdbc:postgresql://ec2-3-9-191-104.eu-west-2.compute.amazonaws.com:5432/testdb \
    --username consultants \
    --password WelcomeItc@2022 \
    --query "SELECT * FROM public.zzzz_seb_table_1 WHERE id > 19 AND \$CONDITIONS" \
    --target-dir /tmp/USUK30/seb/hive/zzzz_seb_table_1/ \
    --split-by id \
    --as-textfile \
    --incremental append \
    --check-column id \
    --last-value 19
```
